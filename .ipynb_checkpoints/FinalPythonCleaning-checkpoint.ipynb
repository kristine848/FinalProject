```python
#import 
import pandas as pd
import os
import numpy as np
```

```python
# file path and read file

file = "acfrData.csv"
data = pd.read_csv(file, encoding='utf8')
data.head(10)
```

```python
#drop columns we don't need
data=data.drop(data.columns[[2,3,4,5,9,10,11,13,14,15,16,18,19,20,21,24,25,26,27,28,31,33,35,37,38,43,44,45,49,50,51,52]], axis = 1)
data.head(10)
```

```python
# new data frame with split value columns 
ITC = data["Basic Incident Type Code And Description (FD1.21)"].str.split(" - ", n = 1, expand = True) 
SubCat = data['Basic Incident Type Subcategory (FD1.21)'].str.split(" - ", n=1, expand =True)
IncidentType = data['Basic Incident Type Category (FD1.21)'].str.split(' - ', n=1, expand = True)
# making separate first name column from new data frame 
data["ITCNumber"]= ITC[0]   
data["IncidentDescription"]= ITC[1] 
data['SubCategoryNumber'] = SubCat[0]
data['SubCategory'] = SubCat[1]
data['IncidentTypeNumber'] = IncidentType[0]
data['IncidentCategory'] = IncidentType[1]
  
# Dropping old Name columns 
data.drop(columns =["Basic Incident Type Code And Description (FD1.21)", "Basic Incident Type Subcategory (FD1.21)","Basic Incident Type Category (FD1.21)"], inplace = True) 
  
# df display 
data 
```

```python
#number of calls by hour grouped by Incident Category
ITCCountByHour = data.groupby(['Hour_of_day', 'ITCNumber', 'IncidentDescription']).ITCNumber.agg('count')
ITCCountByHour
```

```python
#number of calls by Sub Category Number
SubCategoryByHour = data.groupby(['Hour_of_day', 'SubCategoryNumber', 'SubCategory']).SubCategoryNumber.agg('count')
SubCategoryByHour
```

```python
#number of calls by Sub Category Number
IncidentCategoryByHour = data.groupby(['Hour_of_day', 'IncidentTypeNumber', 'IncidentCategory']).IncidentTypeNumber.agg('count')
IncidentCategoryByHour
```
